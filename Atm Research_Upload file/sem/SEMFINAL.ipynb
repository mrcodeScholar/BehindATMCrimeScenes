```python
import semopy as sem
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('Dataset.csv')
# pd.set_option('display.max_columns', None) # Changes dataframe display setting
# Display the DataFrame

data = df[['mway_dist', 'police_dis', 'com_dens','density','unp_rate','age','income','parish','success_attack','freq_attack']]
data
# data = df[['mway_dist', 'police_dis', 'com_dens','density','unp_rate','age','income','success_attack']]
# data
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values per column:")
print(missing_values)

# Check for outliers using summary statistics
summary_stats = data.describe()
print("\nSummary statistics:")
print(summary_stats)

# Check for multicollinearity using correlation matrix
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
formula = """
success_attack ~ com_dens + age + income + unp_rate + density + police_dis + mway_dist + parish

"""

model = sem.Model(formula)
fit = model.fit(data, obj="DWLS", solver="SLSQP",clean_slate=True)

print(sem.calc_stats(model).T)
print(fit)
print(model.inspect())
sem.semplot(model, "success.png", plot_covs=True, std_ests=True, show=True)
```

```python
formula = """


l1 =~ parish
success_attack ~ l1 + mway_dist + police_dis
l1 ~ com_dens + age + income + unp_rate + density


"""

model = sem.Model(formula)

fit = model.fit(data, obj="GLS", solver="SLSQP")  # GLS  WLS  
# sem.bias_correction(model, n=50, resample_mean=True)
print(sem.calc_stats(model).T)
print(fit)
print(model.inspect())
sem.semplot(model, "semresult2.png", plot_covs=True, std_ests=True, show=True)
```

```python
formula = """
L1 =~  age + police_dis + mway_dist + com_dens 
L2 =~ income + unp_rate + density 
success_attack ~ L1 + L2

"""

model = sem.Model(formula)

fit = model.fit(data, obj="DWLS", solver="SLSQP")  #  DWLS 
sem.bias_correction(model, n=500, resample_mean=True)
print(sem.calc_stats(model).T)
print(fit)
print(model.inspect())
sem.semplot(model, "semresult.png", plot_covs=True, std_ests=True, show=True)
```

```python
formula = """


l1 =~ parish
success_attack ~ l1 + mway_dist + police_dis
freq_attack ~ l1 + parish + mway_dist + police_dis
l1 ~ com_dens + age + income + unp_rate + density


"""

model = sem.Model(formula)

fit = model.fit(data, obj="DWLS", solver="SLSQP")  #  DWLS 
sem.bias_correction(model, n=500, resample_mean=False)
print(sem.calc_stats(model).T)
print(fit)
print(model.inspect())
sem.semplot(model, "semresultbias.png", plot_covs=True, std_ests=True, show=True)
```
