```python
import semopy as sem
import pandas as pd
import numpy as np

```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('Dataset.csv')
# pd.set_option('display.max_columns', None) # Changes dataframe display setting
# Display the DataFrame
df.shape
df
```

```python

data = df[['mway_dist', 'police_dis', 'com_dens','density','unp_rate','age','income','success_attack']]
data
```

```python
data.describe()
```

```python
data.corr()
```

```python
# Assuming you have your dataset ready and loaded into a variable named 'data'
formula = """
D =~ com_dens + age + income + unp_rate + density 
G =~  police_dis + mway_dist
success =~ success_attack
success ~  G + D 
"""

model = sem.Model(formula)
model.fit(data, obj="MLW", solver="SLSQP") #MLW ULS GLS FIML DWLS WLS
# sem.bias_correction(model, n=500, resample_mean=False)
print(sem.calc_stats(model).T)
print(model.inspect())
sem.semplot(model, "sem.pdf", plot_covs=True, std_ests=True)
```

```python
formula = """
D =~ com_dens + age + income + unp_rate + density 
G =~  police_dis + mway_dist
success =~ success_attack
success ~  G 
success ~ D
"""
model = sem.Model(formula)
model.load_dataset(data)
opt = sem.Optimizer(model)
objective_function_value = opt.optimize()

# print('ULS ',opt.optimize('ULS'))
opt.optimize('MLW')
model.fit(data, obj="DWLS", solver="SLSQP") #MLW ULS GLS FIML DWLS WLS
print(sem.calc_stats(model).T)
sem.semplot(model, "sem.pdf", plot_covs=True, std_ests=True)
```
